```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
from math import *
import seaborn as sns
sns.set()

def visualize_dataset(X, T, Y=None):
    m = ["x", "o"]  # Indicates true class
    col = ["red", "blue"]  # Indicates prediction
    fig = plt.figure(); plt.axis('equal')
    if Y is not None:
        idx_tp = np.logical_and(T >= .5, Y >= .5) # True positives
        idx_fp = np.logical_and(T < .5, Y >= .5) # False positives
        idx_fn = np.logical_and(T >= .5, Y < .5) # False negatives
        idx_tn = np.logical_and(T < .5, Y < .5) # True negatives
        
        plt.scatter(X[idx_tp,0], X[idx_tp,1], marker=m[0], c="red", label="TP")
        plt.scatter(X[idx_fp,0], X[idx_fp,1], marker=m[1], c="red", label="FP")
        plt.scatter(X[idx_fn,0], X[idx_fn,1], marker=m[0], c="blue", label="FN")
        plt.scatter(X[idx_tn,0], X[idx_tn,1], marker=m[1], c="blue", label="TN")

    else:
        idx_pos, idx_neg = (T > .5), (T < .5)
        plt.scatter(X[idx_pos,0], X[idx_pos,1], marker=m[0], color="blue", label="Pos")
        plt.scatter(X[idx_neg,0], X[idx_neg,1], marker=m[1], color="blue", label="Neg")
    
    plt.xlabel("x"); plt.ylabel("y")
    plt.legend(bbox_to_anchor=(0, 1), loc='upper left', ncol=1)
    return fig.axes[0]

def relu(x):
    return np.maximum(0, x)

def logistic(x):
    return 1 /(1 + np.exp(-x))

def calculate_theta(X, y):
    
    m = X.shape[0] # Number of training examples. 
    # Appending a cloumn of ones in X to add the bias term.
    X = np.append(X, np.ones((m,1)), axis=1)    
    # reshaping y to (m,1)
    y = y.reshape(m,1)
    
    #X = relu(X)
    # The Normal Equation
    theta = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    
    return theta

def predict(X, theta):
    
    # Appending a cloumn of ones in X to add the bias term.
    X = np.append(X, np.ones((X.shape[0],1)), axis=1)
    
    # preds is y_hat which is the dot product of X and theta.
    #preds = np.dot(X, theta)
    
    return X @ theta

def split_dataset(X, T, train=.8):
    N = X.shape[0]
    N_train = int(round(N * train))
    N_test = N - N_train

    X_train, X_test = X[:N_train,:], X[N_train:,:]
    T_train, T_test = T[:N_train], T[N_train:]
    return X_train, T_train, X_test, T_test


# Simple dataset
def generate_dataset(whatType, show = False):
    if whatType == 1:
        n_samples = 100
        n_features = 1
        noise = 20
        X, y = datasets.make_regression(n_samples=n_samples, n_features=n_features, n_informative=1, noise=noise, random_state=37)
        if show == True:
            fig = plt.figure(figsize=(4, 4))
            plt.xlabel("x (feature)")
            plt.ylabel("y (output)")
            plt.title("Synthetic data set")
            plt.scatter(X, y)
            plt.show()
    if whatType == 2:
        # Complex dataset
        n_samples = 300
        x = np.linspace(-10, 10, n_samples) # coordinates
        noise_sample = np.random.normal(0,0.5,n_samples)
        sine_wave = x + np.sin(4*x) + noise_sample
        plt.plot(x, sine_wave, 'o');
        plt.show()
    
    return X, y

def calculate_error_4(X, y, theta):
    X = np.append(X, np.ones((X.shape[0],1)), axis=1)
    
    N = len(X)
    #print(N, n)
    
    preds = np.dot(X, theta)

    err = (1/N) * np.sum(np.power(np.subtract(preds , y), 2))
    return err;
    
def calculate_error_5(X, y, theta):
    return sqrt(calculate_error_4(X, y, theta))

def calculate_error_gradient(X, y, theta):
    
    #X = np.append(X, np.ones((X.shape[0],1)), axis=1)

    N = len(X)
    preds = X @ theta
    err = (1/N) * np.sum(np.power(np.subtract(preds , y), 2))
    return sqrt(err)

def gradient_descent(X, y, lr=0.01, epochs_no = 10, mode = 1):
    (m, D) = X.shape
    X = np.append(X, np.ones((m,1)), axis=1)    
    
    theta = np.random.randn((D+1))
    epochs = epochs_no;
    X_train, y_train, X_test, y_test = split_dataset(X, y)

    train_acc, test_acc = [], []
    train_lr = [[],[],[],[],[]]
    
    lr_list = [0.001, 0.01, 0.1, 0.2, 0.5]
    theta_list = [theta, theta, theta, theta, theta]
    
    if mode == 1:
        for k in range(epochs_no):
            theta -= lr * (np.transpose(X) @ ((X @ theta) - y)) / m

            #T_train = X_train @ theta
            #T_test = X_test @ theta
            train_acc.append(calculate_error_gradient(X_train, y_train, theta))
            test_acc.append(calculate_error_gradient(X_test, y_test, theta))
    else:
        for i in range(len(lr_list)):
            for k in range(epochs_no):
                train_lr[i].append(calculate_error_gradient(X_train, y_train, theta_list[i]))
                theta_list[i] -= lr_list[i] * (np.transpose(X) @ ((X @ theta_list[i]) - y)) / m
                
                
    epochs_no = len(train_acc)
    
    if mode == 1:
        fig, ax1 = plt.subplots(1,1);
        print(train_acc, range(epochs_no))

        ax1.plot(range(epochs_no), train_acc, 'go-', label="Train Loss Value")
        ax1.plot(range(epochs_no), test_acc, 'ro-', label="Test Loss Value")
        ax1.set_ylabel("empirical risk")
        ax1.set_xlabel("number of iterations")
        ax1.legend(loc='upper right', ncol=1)
    else:
#         fig, ax2 = plt.subplots(1,1);
#         print(train_lr[0], range(epochs_no))
#         ax2.plot(range(epochs_no), train_lr[0], 'go-', label="Learning rate = 1.00e-03")
#         ax2.plot(range(epochs_no), train_lr[1], 'bo-', label="Learning rate = 1.00e-02")
#         ax2.plot(range(epochs_no), train_lr[2], 'ro-', label="Learning rate = 1.00e-01")
#         ax2.plot(range(epochs_no), train_lr[3], 'mo-', label="Learning rate = 2.00e-03")
#         ax2.plot(range(epochs_no), train_lr[4], 'wo-', label="Learning rate = 5.00e-03")
#         ax2.set_ylabel("empirical risk")
#         ax2.set_xlabel("number of iterations")
#         ax2.legend(loc='upper right', ncol=1)
        fig = plt.figure(figsize=(4, 4))
        plt.xlabel("x (feature)")
        plt.ylabel("y (output)")
        plt.title("Synthetic data set")
        print(train_lr[0], range(epochs_no))
        plt.plot(range(epochs), train_lr[0], 'go-', label="Learning rate = 1.00e-03")
        plt.plot(range(epochs), train_lr[1], 'bo-', label="Learning rate = 1.00e-02")
        plt.plot(range(epochs), train_lr[2], 'ro-', label="Learning rate = 1.00e-01")
        plt.plot(range(epochs), train_lr[3], 'mo-', label="Learning rate = 2.00e-03")
        plt.plot(range(epochs), train_lr[4], 'wo-', label="Learning rate = 5.00e-03")
        plt.legend(bbox_to_anchor=(1.0, 1.05))
        plt.show()
        
    return theta
    
def problem_1_1():
    print()
    print("REZULTATE PROBLEM 1.1")
    print()
    
    X, y = generate_dataset(1)
    
    theta = calculate_theta(X, y)
    print("Theta =", theta)
    
    preds = predict(X, theta)
    print(preds)
    # Plotting the predictions.
    fig = plt.figure(figsize=(4, 4))
    plt.xlabel("x (feature)")
    plt.ylabel("y (output)")
    plt.title("Synthetic data set")
    plt.plot(X, y, 'co')
    plt.plot(X, preds, 'r-')
    plt.show()

    # Plotting the predictions.
    fig = plt.figure(figsize=(4, 4))
    plt.xlabel("true output")
    plt.ylabel("predicted output")
    plt.title("Quality of predictions")
    plt.plot(y, preds, 'co')
    plt.show()
    
    X_train, y_train, X_test, y_test = split_dataset(X, y)
    
    theta = calculate_theta(X_train, y_train)
    print("Theta =", theta)
    
    preds = predict(X_train, theta)    
    
    print("Eroarea training - (4) =",calculate_error_4(X_train, y_train, theta))
    print("Eroarea training - (5) =",calculate_error_5(X_train, y_train, theta))
    
    theta = calculate_theta(X_test, y_test)
    print("Theta =", theta)
    
    preds = predict(X_test, theta)    
    
    print("Eroarea testare - (4) =",calculate_error_4(X_test, y_test, theta))
    print("Eroarea testare - (5) =",calculate_error_5(X_test, y_test, theta))

def problem_1_2():
    print()
    print("REZULTATE PROBLEM 1.2")
    print()

    X, y = generate_dataset(1)
    theta = gradient_descent(X, y, lr=.05, epochs_no=10, mode = 1)
    
    print("Eroarea training - (4) =", calculate_error_4(X, y, theta))
    print("Eroarea training - (5) =", calculate_error_5(X, y, theta))
    
    print(theta)
    
    Y = predict(X, theta)
        
    theta_closed_form = calculate_theta(X, y)
    print("Theta =", theta)
    
    preds = predict(X, theta_closed_form)
    # Plotting the predictions.
    fig = plt.figure(figsize=(4, 4))
    plt.xlabel("x (feature)")
    plt.ylabel("y (output)")
    plt.title("Synthetic data set")
    plt.plot(X, y, 'co')
    plt.plot(X, Y, 'm-')

    plt.plot(X, preds, 'r-')
    plt.show()

    # Plotting the predictions.
    fig = plt.figure(figsize=(4, 4))
    plt.xlabel("true output")
    plt.ylabel("predicted output")
    plt.title("Quality of predictions")
    plt.plot(y, preds, 'co')
    plt.plot(y, Y, 'mo')
    plt.show()
    
    gradient_descent(X, y, lr=.05, epochs_no=10, mode = 0)
    
problem_1_1()
problem_1_2()

```
